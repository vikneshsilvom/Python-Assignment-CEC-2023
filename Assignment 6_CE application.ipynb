```python
# Function to calculate mass flow rate
def calculate_mass_flow_rate(density, flow_rate):
    # Calculate mass flow rate using the formula m = ρ * V
    return density * flow_rate

# Function to calculate pump power
def calculate_pump_power(mass_flow_rate, g, height_difference):
    # Calculate pump power using the formula W = m * g * h
    return mass_flow_rate * g * height_difference

# Get data for the pipe
pipe_name = input("Enter the name of the pipe: ")

# Get data for mass flow rate
print("Data for mass flow rate:")
density = float(input("Enter the density of the fluid in kg/m³: "))
flow_rate = float(input("Enter the flow rate in m³/s: "))

# Calculate mass flow rate
mass_flow_rate = calculate_mass_flow_rate(density, flow_rate)

# Get data for pump power calculation
print("\nData for pump power calculation:")
g = float(input("Enter the gravitational acceleration in m/s²: "))
height_difference = float(input("Enter the height difference in meters: "))

# Calculate pump power
pump_power = calculate_pump_power(mass_flow_rate, g, height_difference)

# Display the results
print("\nResults:")
print(f"Pipe Name: {pipe_name}")
print(f"Mass Flow Rate: {mass_flow_rate} kg/s")
print(f"Pump Power Required: {pump_power} Watts")
```
